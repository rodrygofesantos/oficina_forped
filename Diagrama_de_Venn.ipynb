### Diagrama de Venn

O **Diagrama de Venn** é uma visualização clássica para representar **interseções, exclusividades e relações entre conjuntos de dados**.

Na **Ciência da Informação**, ele é especialmente útil para:

- Comparar **bases de dados** (ex: repositórios institucionais, catálogos, arquivos),
- Identificar **registros duplicados ou exclusivos**,
- Apoiar **auditorias informacionais**,
- Controlar **versões de documentos** e validar fontes de origem.

Ao usar o Diagrama de Venn, conseguimos **visualizar sobreposições** de forma clara, facilitando diagnósticos e decisões sobre integração, interoperabilidade e limpeza de dados.

Nesta seção, aprenderemos a gerar diagramas de Venn com Python utilizando a biblioteca `matplotlib_venn`.


```python
#importar as bibliotecas
import matplotlib.pyplot as plt
from matplotlib_venn import venn2, venn3
```

# Exemplo 1: Duas bases bibliográficas
**Cenário:**
*   Base A: artigos indexados até 2010.
*  Base B: artigos indexados de 2008 a 2015.

**Dados:**
*   300 artigos apenas em A.
*   400 apenas em B.
*   120 em comum..


```python
venn2(subsets=(300, 400, 120), set_labels=("Base A", "Base B"))
plt.show()
```

**Exemplo 2: Três repositórios institucionais**

A distribuição de registros entre três acervos digitais: Biblioteca Central (BC), Repositório Institucional (RI) e Base de Teses e Dissertações (BTD).
Cada região do diagrama de Venn revela relações específicas:


* A (100): Registros exclusivos da BC, indicando conteúdo singular e não integrado.

* B (200): Registros exclusivos da RI, sugerindo centralidade na produção acadêmica.

* C (0): Ausência de compartilhamento direto entre BC e RI sem a BTD.

* D (0): A BTD não possui registros próprios, o que questiona sua autonomia.

* E (0): Não há interseção direta entre BC e BTD, sugerindo baixa interoperabilidade.

* F (120): Registros compartilhados entre RI e BTD, apontando dependência da BTD em relação à RI.

* G (0): Ausência de registros presentes simultaneamente nas três bases, indicando falta de sincronização total.

```python
# Subconjuntos na ordem: A, B, C, D, E, F, G
# A = apenas BC
# B = apenas RI
# C = BC e RI (sem BTD)
# D = apenas BTD
# E = BC e BTD (sem RI)
# F = RI e BTD (sem BC)
# G = BC, RI e BTD

# Criar a figura e ajustar o tamanho (opcional)
fig = plt.figure(figsize=(8, 6))

venn3(subsets=(100, 200, 0, 0, 0, 120, 0),
      set_labels=('Biblioteca Central', 'Repositório Institucional', 'Base de Teses e Dissertações'))

# Título com maior espaçamento do diagrama
plt.suptitle("Comparação de Registros entre Três Acervos Digitais", fontsize=14, y=0.95)

plt.tight_layout()
plt.show()
```
