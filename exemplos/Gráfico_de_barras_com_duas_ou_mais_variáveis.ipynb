### Gráfico de Barras com Duas ou Mais Variáveis

O gráfico de barras com múltiplas variáveis permite **comparar diferentes categorias simultaneamente**, facilitando a visualização de **relações cruzadas** em um mesmo eixo.

Na Ciência da Informação, esse tipo de gráfico é útil para responder perguntas como:

- Quantos documentos foram produzidos por tipo e por ano?
- Como se distribuem os projetos por grupo de pesquisa e situação (em execução ou finalizado)?
- Qual o número de autores por curso e por instituição?

Esse tipo de visualização é ideal para **identificar padrões e contrastes** entre variáveis relacionadas, principalmente em bases informacionais com metadados variados.

Nesta etapa, aprenderemos a construir gráficos de barras agrupadas ou lado a lado usando `pandas` e `matplotlib`, utilizando dados reais ou simulados.


```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('/content/grafico_de_barras_com_duas_ou_mais_variaveis.xlsx')
```

```python
df.sample(10)
```

```python
# Ordena em ordem decrescente a coluna 'Nomination'

df.sort_values(by='Nomination', ascending=True, inplace=True)
```

```python
# Imprime na tela as cinco primeiras linhas

df.head()
```

```python
# Cria lista com os títulos dos filmes

filmes = df['Film'].to_list()
```

```python
# Cria lista com total de indicações

total_indicacao = df['Nomination'].to_list()
```

```python
# Cria lista com total de prêmios

total_premio = df['Award'].to_list()
```

```python
fig, ax = plt.subplots()

bar_container_indicacao = ax.bar(filmes, total_indicacao, color='purple')
bar_container_premiacao = ax.bar(filmes, total_premio, bottom=total_indicacao, color='green')
ax.bar_label(bar_container_indicacao, label_type='center')
ax.bar_label(bar_container_premiacao, label_type='center')
plt.xticks(rotation = 90)

ax.set_ylabel('Total de indicações')
ax.set_title('Indicados para o Oscar 2022')

plt.show()

```

```python
# Criando visualização a partir do Pandas

df.plot(
    x='Film',
    y=['Award', 'Nomination'],
    title='Oscar 2022',
    ylabel='Quantidade',
    color=['olive', 'purple'],
    kind='bar')

plt.show()
```
